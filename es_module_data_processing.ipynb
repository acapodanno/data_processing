** Punto 4.**

```python
df_drop_columns = df.copy()
num_to_remove_column = int(df_drop_columns.shape[0]*0.7)
df_drop_columns.dropna(axis=1,thresh=num_to_remove_column)
```

**Punto 5.**

```python
df_drop_rows = df.copy()
num_to_remove_row = int(df_drop_rows.shape[1]*0.75)
df_drop_rows.dropna(axis=0,thresh=num_to_remove_row)
```

** Punto 6.**

```python
df_drop_column_price = df.copy()
df_drop_column_price.dropna(axis=0,subset=['PRICE'])
df_drop_column_price.head()
```

```python
df_drop_column_price.shape
```

**Punto 7**

```python
cols = df.columns
cols = cols.drop(["CRIM","CHAS"])
cols
```

```python
replace_with = df[cols].mean()
df = df.fillna(replace_with)
df.head()
```

**Punto 8**

```python
!pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder
df_label_encoder = df.copy()
le = LabelEncoder()
df_label_encoder["CHAS"] = le.fit_transform(df_label_encoder["CHAS"])
df_label_encoder.head()
```

```python
df_ordinal_encoder = df.copy()
size_mapping = {"LOW":1,"MODERATE":2,"HIGH":3,"VERY HIGH":4}
df_ordinal_encoder["CRIM"] = df_ordinal_encoder["CRIM"].map(size_mapping)
df_ordinal_encoder.head()
```

**Punto 9**

```python
df = pd.get_dummies(df,columns=["CRIM"])
df.head()
```

**Punto 10**

```python
from sklearn.preprocessing import StandardScaler

df_std = df.copy()
features = ["INDUS","LSTAT","RAD","B","PRICE"]
to_std = df_std[features]
scaler = StandardScaler()
df_std[features] = scaler.fit_transform(to_std)
df_std.head()
```

**Punto 11**

```python
df_std.to_csv("housing_clean.tsv", sep='\t', encoding='utf-8', index=False, header=True)
```
