**Feature Scaling - Normalizzazione**

Significa transformare le feauture con un valore compresso tra 0 e 1.

**Formula**:
xi = xi - min(x)/ max(x) - min(x)

**Pandas**

```python
import pandas as pd
url = "https://raw.githubusercontent.com/ProfAI/machine-learning-fondamenti/main/datasets/wine.csv"
df = pd.read_csv(url, usecols=[0,1,7])
df.head()

```

```python
wine_norm = df.copy()
features = ["alcol","flavonoidi"]
to_norm = wine_norm[features]

wine_norm[features] = (to_norm - to_norm.min())/(to_norm.max() - to_norm.min())
wine_norm.head()
```

**Verifica Normalizzazione** Per verificare se la normalizzazione Ã¨ corretta prendere il min e max che devono corrispondere a 0 e 1.

```python
wine_norm.min()
```

```python
wine_norm.max()
```
