###Standardizzazione
Distribuzione normalizzata media a zero e devazione a 1

**Formula**: xstd = (xi - mean(x))/std(x)

###Pandas

```python
import pandas as pd
url = "https://raw.githubusercontent.com/ProfAI/machine-learning-fondamenti/main/datasets/wine.csv"
wine = pd.read_csv(url, usecols=[0,1,7])
wine.head()
```

```python
wine_std = wine.copy()
features = ["alcol","flavonoidi"]
to_std= wine_std[features]
wine_std = (to_std - to_std.mean())/to_std.std(ddof=0)
wine_std.head()
```
